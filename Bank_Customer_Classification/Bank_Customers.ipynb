```python
x_geo = pd.get_dummies(x['Geography'], drop_first=True)
x_geo
```

```python
x_gen = pd.get_dummies(x['Gender'], drop_first=True)
x_gen
```

```python
x
```

```python
x = pd.concat([x_geo, x], sort=True, axis=1)
```

```python
x = pd.concat([x_gen, x], sort=True, axis=1)
x
```

```python
x.drop(['Geography','Gender'], inplace=True, axis=1)
```

```python
x = x.values
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
import tensorflow as tf
```

```python
import keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(output_dim= 6, init= 'uniform', activation= 'relu', input_dim= 11))
```

```python
classifier.add(Dense(output_dim= 6, init= 'uniform', activation= 'relu'))
```

```python
classifier.add(Dense(output_dim= 1, init= 'uniform', activation= 'sigmoid'))
```

```python
classifier.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
classifier.fit(x_train, y_train, batch_size=10, epochs=100)
```

```python
y
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
